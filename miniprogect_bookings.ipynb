```python
#Приведём названия колонок к нижнему регистру
book.columns=book.columns.str.lower()
```

```python
#Заменим пробелы на знак нижнего подчеркивания
book.columns=book.columns.str.replace(' ','_')
```

```python
book
```

```python
#Посмотрим пользователи из каких стран совершили наибольшее число успешных бронирований
usp=book.query('is_canceled==0') \
    .groupby('country') \
    .aggregate({'is_canceled':'count'}) \
    .sort_values('is_canceled')
```

```python
usp
```

```python
#На сколько ночей в среднем бронируют отели разных типов
nig=book.query("hotel=='City Hotel'") \
    .aggregate({'stays_total_nights':'mean'})
```

```python
nig
```

```python
gin=book.query("hotel=='Resort Hotel'") \
    .aggregate({'stays_total_nights':'mean'})
```

```python
gin
```

```python
#Иногда тип номера, полученного клиентом, отличается от изначально забронированного. Проверим сколько таких случаев.
hi=book.query('assigned_room_type!=reserved_room_type') \
    .aggregate({'assigned_room_type':'count'})
```

```python
hi
```

```python
#На какой месяц чаще всего успешно оформляли бронь в 2016
lh=book.query('arrival_date_year==2016') \
    .groupby('arrival_date_month') \
    .aggregate({'is_canceled':'count'}) \
    .sort_values('is_canceled')
```

```python
lh
```

```python
#На какой месяц чаще всего успешно оформляли бронь в 2017
hl=book.query('arrival_date_year==2017') \
    .groupby('arrival_date_month') \
    .aggregate({'is_canceled':'count'}) \
    .sort_values('is_canceled')
```

```python
hl
```

```python
#Проверим, на какой месяц бронирования отеля типа City Hotel отменялись чаще всего в каждый из периодов
kl=book.query("hotel=='City Hotel' & is_canceled==1") \
    .groupby(['arrival_date_year', 'arrival_date_month']) \
    .aggregate({'is_canceled':'count'}) 
```

```python
total_kids=book.children+book.babies
```

```python
#Создадим колонку total_kids, объединив children и babies.
book=book.assign(total_kids=book.children+book.babies)
```

```python
book
```

```python
#Посмотрим для отелей какого типа среднее значение переменной оказалось наибольшим
jk=book.query("hotel=='City Hotel'") \
    .aggregate({'total_kids':'mean'})
```

```python
jk
```

```python
kj=book.query("hotel=='Resort Hotel'") \
    .aggregate({'total_kids':'mean'})
```

```python
kj
```

```python
k=book.agg({'is_canceled':'count'})
g=book.query('total_kids>=1') \
    .agg({'is_canceled':'count'})
l=book.query('is_canceled==1 & total_kids>=1') \
    .agg({'is_canceled':'count'})
p=l/g
print(k)
print(g)
print(l)
print(p)
```

```python
c=book.agg({'is_canceled':'count'})
u=book.query('total_kids<1') \
    .agg({'is_canceled':'count'})
b=book.query('is_canceled==1 & total_kids<1') \
    .agg({'is_canceled':'count'})
f=b/u
print(c)
print(u)
print(b)
print(f)
```
